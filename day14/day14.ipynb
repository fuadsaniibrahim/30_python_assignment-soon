#LEVEL 1

##map Function:
__Purpose__: map applies a given function to all items in an iterable (e.g., a list) and returns an iterable of the results.
__Syntax__: map(function, iterable)
###Example:
code
numbers = [1, 2, 3, 4]
squared = map(lambda x: x**2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
__Note__: map returns an iterator, so it is common to wrap it with list() to get the results as a list.


#filter Function:
__Purpose__: filter filters elements from an iterable based on a specified function that returns True or False.
__Syntax__: filter(function, iterable)
###Example:
 code
numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))  # Output: [2, 4, 6]
__Note__: Like map, filter returns an iterator, so it is often used with list().


##reduce Function:
__Purpose__: reduce applies a binary function cumulatively to the items in an iterable to reduce it to a single value.
__Syntax__: functools.reduce(function, iterable[, initializer])
###Example:
code
from functools import reduce
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24
__Note__: While map and filter return iterables, reduce returns a single value. The initializer argument is optional and provides an initial value for the accumulator.


##Higher-Order Function:

**Purpose**: A higher-order function is a function that takes one or more functions as arguments or returns a function as its result. It enables the composition of functions, making code more modular and flexible.

code
def apply_operation(func, x, y):
    return func(x, y)

def add(x, y):
    return x + y

result = apply_operation(add, 3, 4)
print(result)  # Output: 7


##Closure:

**Purpose**: A closure is a function object that has access to variables in its lexical scope, even when the function is called outside that scope. It allows functions to retain state and create private variables.

code
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function

closure_instance = outer_function(5)
result = closure_instance(3)
print(result)  # Output: 8


##Decorator:

**Purpose**: A decorator is a higher-order function that takes a function and extends or modifies its behavior without modifying its source code directly. Decorators are often used to add functionality, logging, or validation to functions.
code
def my_decorator(func):
    def wrapper(*args, **kwargs):
        # Code before calling the decorated function
        result = func(*args, **kwargs)
        # Code after calling the decorated function
        return result
    return wrapper

@my_decorator
def my_function():
    # Function's code
    pass

__Note__: The @my_decorator syntax is a convenient way to apply the decorator to my_function.

Using higher-order functions, closures, and decorators promotes modularity, code reusability, and clean code practices in Python programming.

```python
languages_pop = {}
for country in countriesdata:
    # Check if 'language' key exists in the current country dictionary
    if 'language' in country:
        for language in country['language']:
            languages_pop[language] = languages_pop.get(language, 0) + country['population']

sorted_languages = sorted(languages_pop.items(), key=lambda item: item[1], reverse=True)

print("Top ten most spoken languages by location:")
for rank, (language, population) in enumerate(sorted_languages[:10], 1):
    print(rank, language, population)
    print(f"{rank}. {language}: {population} speakers")

        
```

```python

def most_spoken_languages(countriesdata, lenght = 10):
    all_languages = [language for country in countriesdata for language in country["languages"]]
    language_counts = {language: all_languages.count(language) for language in set(all_languages)}
    sorted_languages = sorted(language_counts.items(), key=lambda x: x[1], reverse=True)
    return sorted_languages[:lenght]

print(most_spoken_languages(countriesdata, 10))
all_languages = [language for country in countriesdata for language in country["languages"]]
```
